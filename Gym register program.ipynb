```python
import operator

members = []
hakan_is_here = False

def name_from_input():
  name = input("Name: ")
  name = name.capitalize()
  if name == 'Hakan':
    global hakan_is_here
    hakan_is_here = True
    name = 'Hasan'
  return name

def surname_from_input():
  surname = input("Surname: ")
  return surname.capitalize()

def age_from_input():
  return int(input("Age: "))

def height_from_input():
  return float(input("Height(m): "))

def weight_from_input():
  weight = float(input("Weight: "))
  global hakan_is_here
  if hakan_is_here: weight += 5
  return weight

def calculate_bmi(height, weight):
  bmi = weight / (height**2)
  return round(bmi, 1)

def calculate_bmi_category(bmi):
  if bmi < 18.5:
    bmi_category = "Underweight"
  elif bmi >= 18.5 and bmi <= 24.9:
    bmi_category = "Normal weight"
  elif bmi >= 25 and bmi <= 29.9:
    bmi_category = "Overweight"
  else:
    bmi_category = "Obesity"
  return bmi_category

def calculate_member(index):
  height = members[index][1]
  weight = members[index][2]
  if members[index][6]: weight += 5 # hakan is here

  bmi = calculate_bmi(height, weight)
  bmi_category = calculate_bmi_category(bmi)

  members[index][4] = bmi
  members[index][5] = bmi_category

def order_members():
  global members
  members = sorted(members, key=operator.itemgetter(3))
  for index, member in enumerate(members):
    if member[6]: # hakan is here
      members.pop(index)
      members.append(member)

def add():
  name = name_from_input()
  surname = surname_from_input()
  age = age_from_input()
  height = height_from_input()
  weight = weight_from_input()

  bmi = calculate_bmi(height, weight)
  bmi_category = calculate_bmi_category(bmi)

  global hakan_is_here
  members.append([name + ' ' + surname, height, weight, age, bmi, bmi_category, hakan_is_here])
  if hakan_is_here: hakan_is_here = False

def remove():
  value = input("Press number for remove user ")
  value = int(value)
  members.pop(value)

def update():
  value = input("Press number for update user ")
  value = int(value)

  age = age_from_input()
  height = height_from_input()
  weight = weight_from_input()

  members[value][1] = height
  members[value][2] = weight
  members[value][3] = age

  calculate_member(value)

def show():
  order_members()
  for index, member in enumerate(members):
    print(f'({index}) {member[0]}, {member[3]} age, {member[1]}m, {member[5]}')

def main():
  actions = {
    'a' : add,
    'r' : remove,
    'u' : update,
    's' : show
  }

  while True:
    value = input("\n(a) Add, (r) Remove, (u) Update, (s) Show\n")
    actions[value]()

if __name__ == '__main__':
  main()
```
